# ${} Significance in Bash

```python

%%bash
# In the below code using {} and not using braces produces the same value

#!/bin/bash
var=" value of var"
echo ${var}

var="value of var"
echo $var

#!/bin/bash
height=170
# Wrong : no such variable called heightcm
echo "Your height is = $heightcm"

# What we want is $height followed by "cm" with no empty string between the two
echo "Your height is = ${height}cm"

pwd
```
